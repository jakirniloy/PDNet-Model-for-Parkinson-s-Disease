# **PROCESSED HC DAT**





```python
# Path to the HC folder containing images
hc_folder_path = "/Capstone_PDorNotDAT/pdOrNotDAT/HC"

# Path to the directory to save processed HC images
processed_hc_folder_path = "/processed_DAT/pdORnot/hc"

# Create the directory if it doesn't exist
os.makedirs(processed_hc_folder_path, exist_ok=True)

count = 0

# Iterate through each image in the HC folder
for image_filename in os.listdir(hc_folder_path):
        image_path = os.path.join(hc_folder_path, image_filename)

        # Load the image
        image = cv2.imread(image_path)

        # Convert the image to grayscale
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

        # Apply threshold to identify regions of high activity
        _, thresholded = cv2.threshold(gray, 190, 255, cv2.THRESH_BINARY)

        # Create a copy of the original image
        highlighted_image = gray.copy()

        # Apply red glow to regions of high activity
        highlighted_image[thresholded > 0] = 255  # Set high activity regions to maximum intensity

        # Convert the highlighted_image to a 3-channel format
        highlighted_image_rgb = cv2.cvtColor(highlighted_image, cv2.COLOR_GRAY2BGR)

        # Apply a different colormap to the highlighted regions for a different color effect
        custom_colormap = cv2.applyColorMap(highlighted_image_rgb, cv2.COLORMAP_HSV )  # Change colormap here

        # Combine the original grayscale image with the custom color effect
        final_image_custom_color = cv2.addWeighted(image, 0.7, custom_colormap, 0.3, 0)

        # Save the processed image with the same name
        processed_image_path = os.path.join(processed_hc_folder_path, image_filename)
        cv2.imwrite(processed_image_path, final_image_custom_color)

        # Display the final image
        #cv2_imshow(final_image_custom_color)

        count = count + 1
        print(count)

print("Processing and saving HC images complete.")
```

```python
import os
import shutil
from sklearn.model_selection import train_test_split
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
```

```python
# Turn images into data loaders
batch_size = 32
NUM_WORKERS = os.cpu_count()

def create_dataloaders(images, labels, transform, batch_size, num_workers=NUM_WORKERS, shuffle=True):
    temp_dir = '/DAT_parkinson_dataset'  # Update to the correct directory
    os.makedirs(temp_dir, exist_ok=True)

    for img_path, label in zip(images, labels):
        class_dir = os.path.join(temp_dir, label)
        os.makedirs(class_dir, exist_ok=True)
        new_path = os.path.join(class_dir, os.path.basename(img_path))
        shutil.copy(img_path, new_path)

    dataset = datasets.ImageFolder(temp_dir, transform=transform)
    class_names = dataset.classes

    #shutil.rmtree(temp_dir)

    dataloader = DataLoader(
        dataset,
        batch_size=batch_size,
        shuffle=shuffle,
        num_workers=num_workers,
        pin_memory=True,
    )

    return dataloader, class_names
```
